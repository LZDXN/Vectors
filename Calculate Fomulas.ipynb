```python
# Original Code by Jing Gao
# Date: Sep. 9th Fri Section 3:30 ~ 4:20

import math
# Unchange conditions
mass = 10 # kg, block
k_spring = 120 # N/m, the spring's stifness
L_0 = 1.0 # m, the spring's relaxed length
g = -10 # m/s^2, the approximate gravitional acceleration constant

# Before the first step
L_now = 2.0 # m, the length of the spring now
v_0 = -1.0 # m/s, the block's initial velocity
p_0 = v_0 * mass # m*kg/s, the block's initial momentum
time_step = 0.02


# the angle of inclination
height = 3 # m
width = 4 # m
theta_rad = 0.6435 # from the result of $tan^{-1}(\frac{3}{4})$
theta_deg = round(theta_rad / math.pi * 180, 4) # transfer the result from radian to degree (if needed)
print("The theta in radian is:" + str(theta_rad) + ", in degree is: " + str(theta_deg) + "Ëš")

# step 1
s = L_0 - L_now
r = -2 # [-2,0,0] the position provided from the length streched of the spring
r_hat = r / L_now # the unit vector of r

# Force
F_spring = s * k_spring * r_hat # the spring force
# print("F_spring:", F_spring) # check point

F_g_horizontal = round(mass * g * math.sin(theta_rad)) # the horizontal component of the gravity
# print("F_g_horizontal:", F_g_horizontal)

F_net = F_spring + F_g_horizontal
# print("F_net:", F_net, "N")

# change of momentum & momentum update
p_f = round(p_0 + F_net * time_step, 4) # round that looks like hand-calculation
p_0 = p_f # update for next step
print("1st step momentum: <", p_f, ", 0 , 0 > m*kg/s")

# change of position & position update
r_f = round(r + (p_f / mass) * time_step, 4)
print("1st step position:<", r_f, ", 0 , 0 > m")
print()

# step 2
L_now = abs(r_f)
s = abs(L_0 - L_now)
r = r_f # [-2,0,0] the position provided from the length streched of the spring
r_hat = r / L_now # the unit vector of r


# Force
F_spring = - s * k_spring * r_hat # the spring force
# print("F_spring:", F_spring) # check point

F_g_horizontal = round(mass * g * math.sin(theta_rad)) # the horizontal component of the gravity
# print("F_g_horizontal:", F_g_horizontal)

F_net = F_spring + F_g_horizontal
# print("F_net:", F_net, "N")

# change of momentum & momentum update
p_f = round(p_0 + F_net * time_step, 3)
print("Final momentum: <", p_f, ", 0 , 0 > m*kg/s")

# change of position & position update
r_f = round(r + (p_f / mass) * time_step, 3)
print("Final position: <", r_f, ", 0 , 0 > m")
```

```python
class Vector():
    def __init__(self, i, j, k):
        self.i = i
        self.j = j
        self.k = k
        self.figure = '.5'
        
    def __repr__(self):
        return f'<{self.i}, {self.j}, {self.k}>'

    def __str__(self):
        return f'<{self.i}, {self.j}, {self.k}>'
    
    def __copy__(self):
        return Vector(self.i, self.j, self.k)
    
    def changeFigure(self, figure):
        self.figure = '.%sg' % figure
        return self
    
    def formating(self, value):
        return float(format(value, self.figure))
    
    def __add__(self, other):
        i = self.formating(self.i + other.i)
        j = self.formating(self.j + other.j)
        k = self.formating(self.k + other.k)
        return Vector(i, j, k)

    def __sub__(self, other):
        i = self.formating(self.i - other.i)
        j = self.formating(self.j - other.j)
        k = self.formating(self.k - other.k)
        return Vector(i, j, k)

    def __iadd__(self, other):
        i = self.formating(self.i + other.i)
        j = self.formating(self.j + other.j)
        k = self.formating(self.k + other.k)
        return Vector(i, j, k)

    def __isub__(self, other):
        i = self.formating(self.i - other.i)
        j = self.formating(self.j - other.j)
        k = self.formating(self.k - other.k)
        return Vector(i, j, k)
    
    def __mul__(self, scalar): # times scalar
        i = self.formating(self.i * scalar)
        j = self.formating(self.j * scalar)
        k = self.formating(self.k * scalar)
        return Vector(i, j, k)
    
    def __truediv__(self, scalar): # dvided scalar
        i = self.formating(self.i / scalar)
        j = self.formating(self.j / scalar)
        k = self.formating(self.k / scalar)
        return Vector(i, j, k)
    
    def __imul__(self, scalar): # times scalar
        i = self.formating(self.i * scalar)
        j = self.formating(self.j * scalar)
        k = self.formating(self.k * scalar)
        return Vector(i, j, k)
    
    def __itruediv__(self, scalar): # dvided scalar
        i = self.formating(self.i / scalar)
        j = self.formating(self.j / scalar)
        k = self.formating(self.k / scalar)
        return Vector(i, j, k)
    
    def dotProduct(self, other):
        i = self.formating(self.i * other.i)
        j = self.formating(self.j * other.j)
        k = self.formating(self.k * other.k)
        return Vector(i, j, k)
    
    def crossProduct(self, other):
        i = self.formating(self.j * other.k - self.k * other.j)
        j = self.formating(-(self.i * other.k - self.k * other.i))
        k = self.formating(self.i * other.j - self.j * other.i)
        return Vector(i, j, k)
    
    def magnitude(self):
        return self.formating((self.i ** 2 + self.j ** 2 + self.k ** 2) ** (1 / 2))
    
    def unit(self): # for unit vector
        return self.__truediv__(self.magnitude())
    
    # projection, theta
        
```
